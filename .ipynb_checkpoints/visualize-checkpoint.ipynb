```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
x = range(20)
y = tf.random.normal([20], 0, 1)

print(y)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'bo')
plt.show()
```

```python
random_normal = tf.random.normal([100000], 0, 1)
```

```python
# hist() = 히스토그램
# 여기서 bins는 100개의 영역에 대하여 도수를 구하여 막대그래프로 나타내는 것을 정의.

plt.hist(random_normal, bins = 100)    
plt.show()
```
